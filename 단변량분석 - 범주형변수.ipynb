# **단변량분석_범주형변수**


## 범주형 변수
- 고유한 값이나 범주 수가 제한된 변수(ex. 혈액형, 결제 방법 등 )
          
- 범주형 변수를 정리하는 방법 -> 범주형 변수는 범주별 빈도수와 비율을 확인한다 <br>
    * 범주별 빈도수
        * 시리즈.value_counts()
    * 범주별 비율
        * 범주별 빈도수를 전체 count로 나누기
        * df.shape : 데이터프레임의 행,열 수를 반환한다
- 시각화
    * Bar plot
	* Pie chart
    
### **1. 수치화 : 기초통계량**

**1-1. 범주별 빈도수**<br>
- .value_counts() : 범주의 개수와 상관 없이 범주 별 개수를 count

   
**1-2. 범주별 비율**
- .value_counts(normalize=True)
    

### **2. 시각화**
**2-1. Bar chart**
- plt.bar() -> 먼저 집계한 후 결과를 가지고 그래프 그려야 한다
- sns.countplot() -> 집계 + bar plot를 한꺼번에 해결한다

**2-2. Pie chart**
- 범주별 비율 비교할 때 파이차트 사용
- 먼저 집계한 후 사용
	* plt.pie( 값, labels=범주이름, autopct = ‘%.2f%%’)
         * autopct = ‘%.2f%%’ : 그래프에 표시할 값 비율 값에 대한 설정
         * .2f% : 소수점 두 자리 퍼센트로 표기 한다는 의미




```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = './SubwayMenuNutrition.csv'
data = pd.read_csv(path)
data.head()
```

```python
data['Category'].value_counts() # 판매하는 음식의 종류별
```

```python
data['Category'].value_counts(normalize = True)
```

```python
# 시각화
sns.countplot(x = 'Category', data = data)
plt.grid()
plt.show()
```

```python
temp = data['Category'].value_counts()

plt.pie(temp.values, labels = temp.index, autopct = '%.2f%%')
plt.show()
```
