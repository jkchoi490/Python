## 숫자형 변수
- 양적 데이터(정량적 데이터)
- 숫자형 변수를 정리하는 두 가지 방법
    * 숫자로 요약하기 : 정보의 대푯값
        * -> 기초 통계량
        * -> min, max, 평균(mean),중앙값(median), 최빈값(mode), 표준편차(std), 사분위수(Quantile)...
    * 구간을 나누고 빈도수(frequency) 계산하기
        * -> 도수분포표(frequency table)
- 시각화
    * 기초 통계량 -> Boxplot
	* 도수분포표 -> Histogram, Desity plot
    
### **1. 수치화 : 대푯값**
**1-1. 평균(산술평균)**<br>
- 평균(mean)
    * 산술 평균 : 모든값들을 더한 후 개수로 나눈 수(합의 평균)
    * 기하 평균 : n개의 양수 값을 모두 곱한 것의 n제곱근(곱의 평균)
    * 조화 평균 : 분자가 동일한 두 비율의 평균
    * ex) np.mean(data['점수']), data['점수'].mean()

**1-2. 중앙값(중위수, median)**
- 자료의 순서상 가운데 위치한 값
    * ex) np.median(data['점수']), data['점수'].meadian()
    
**1-3. 최빈값(mode)**
- 자료 중에서 가장 빈번하게 나타나는 값(빈도가 높은 값)
- 이산형 데이터(셀 수 있는 데이터)
    * data['점수'].mode()
    

**1-4. 사분위수**
- 데이터를 4등분 한 것 (3개 Q1, Q2, Q3)
       ->  (25%, 50%, 75%) 각 경계에 해당되는 값
- 상위 75%(Q1) < 상위 50%(Q2) < 상위 25%(Q3)
- 하위 25%(Q1) < 하위 50%(Q2) < 하위 75%(Q3)
- 자료를 크기 순으로 배열하여 누적 백분율을 4 등분하는 3개 지점에 해당하는 값 ( Q1, Q2, Q3 )
- Q2 는 중앙값과 같다

### **3. 시각화**
**3-1. 히스토그램(Histogram)**
- sns.histplot
- plt.hist(변수명, bins=구간 수) -> bins를 적절히 조절해야 함(구간의 개수에 따라 파악할 수 있는
    내용이 달라짐)
- edgecolor : bar의 윤곽선을 회색으로 표시
- 히스토그램을 저장하면 튜플
- 튜플 안에는 Numpy Array가 2개로 구성된다
	* 첫번째  :  빈도수
	* 두번째  :  구간 값


**3-2. 밀도함수 그래프(Density Plot - kde plot)**
* 히스토그램의 단점: 구간(bin)의 너비를 어떻게 잡는지에 따라 전혀 다른 모양이 될 수 있음
- 밀도함수 그래프
	* 구간의 너비를 정하지 않아도 된다
    * 막대의 너비를 가정하지 않고 데이터의 밀도를 추정하는 커널 밀도 추정(kernel Density Estimation) 방식 사용하여 히스토그램의 단점을 해결한다
    * 면적으로 구간에 대한 확률을 추정한다
    * sns.kdeplot(변수명)
    * 밀도 추정 : 측정(관측)된 데이터로부터 전체 데이터 분포의 특성을 추정한다(ex. 마트의 1년 일일 판매량을 측정하여 kde plot을 그리고 특정한 날의 판매량이 얼마나 될지 확률로 추정)
    

**3-3. boxplot**
- 데이터에 이상치가 얼마나 포함되어 있는지 쉽게 판단할 수 있다
- plt.boxplot() 
    * 값에 NaN이 있으면 그래프가 그려지지 않는다 
    * ex) plt.boxplot(temp['점수'], vert=False) -> vert 옵션 : 횡(False), 종(True, 기본값)
- sns.boxplot()
    * NaN값을 알아서 제거해 준다

![](https://www.simplypsychology.org/wp-content/uploads/box-whisker-plot.jpg) 

* Box(박스)
    - 중앙값을 기준으로 상하위 25% 지점에 있는 데이터의 범위를 표시
    - 전체 데이터를 오름차순으로 정렬한 후 전체 데이터의 25% 지점은 1사분위수(Q1), 75% 지점은 3사분위수(Q3)를 나타낸다
    - 사분위수
        * Q1 : Lower quartile -> 1사분위수
        * Q2 : median  -> 2사분위수
        * Q3 : Upper quartile -> 3사분위수
        
* median(중앙값)
    - 2사분위수(Q2)
    - 2사분위수를 기준으로 데이터의 50%가 상위에, 50%가 하위에 분포되어 있음을 나타낸다
    
* whisker(수염)
    - 상자의 양 끝과 연결된 선
    - 상자길이(IQR)의 1.5배만큼 떨어진 지점을 나타낸다
    - 1사분위수에서 상자길이의 1.5배 떨어진 거리 안에 있는 데이터의 최댓값(Maximum -> 위쪽)과 최솟값(Minimum - 아래쪽)을 기준으로 수염을 표시
    
* IQR(사분범위) -> Inter Quartile Range : 
    - 3분위수 - 1사분위 수
    
![](https://scc.ms.unimelb.edu.au/__data/assets/image/0011/3514655/Trump2.png)

* Actual Whisker Length 
    - 1.5*IQR 범위 이내의 최소, 최대값으로 결정한다
    
    
* Potential Whisker Length
    - 1.5*IQR 범위, 잠재적 수염의 길이 범위
    
    
* Outlier(이상점)
     - 수염 밖에 표시한 데이터
     
     
* Extreme(극단점)
     - 이상점 중에서도 가장 바깥쪽에 위치한 데이터
     - 이상치 가운데서도 유달리 높거나 낮은 값을 보이는 특이한 데이터
